8) Создать матрицу с числами в диапазоне от 1 до 4 по диагонали. В других ячейках матрицы должны быть семерки.

```python
matrix = [[7 for _ in range(4)] for _ in range(4)]  # Создаем матрицу 4x4 заполненную семерками
for i in range(4):
    matrix[i][i] = i + 1  # Заполняем числами от 1 до 4 по диагонали
    # Выводим матрицу на экран
for row in matrix:
    print(row)
```

9) Создать матрицу размера 5 на 5 заполненную чилами от 0 до 4 в строках.

```python
matrix = np.zeros((5, 5), dtype=int)  # Создаем матрицу из нулей размером 5 на 5
for i in range(5):  # Проходим по каждой строке матрицы
    matrix[i] = np.arange(5)  # Заполняем текущую строку числами от 0 до 4
print(matrix)
```

10) Создать матрицу случайных чисел размера 10 на 10 и отнять среднее по строкам из каждого элемента строки

```python
import numpy as np
matrix = np.random.rand(10, 10) # Создание матрицы размера 10 на 10 со случайными числами от 0 до 1
row_means = np.mean(matrix, axis=1) # Вычисление среднего по каждой строке
matrix -= row_means.reshape(-1, 1) # Вычитание среднего из каждого элемента строки
print(matrix)
```

11) Создать матрицу случайных чисел размера 10 на 10 и отнять медиану по столбцам из каждого элемента столбца  

```python
import numpy as np
matrix = np.random.rand(10, 10) # Создание матрицы размера 10 на 10 со случайными числами от 0 до 1
column_medians = np.median(matrix, axis=0) # Вычисление медианы по столбцам
result_matrix = matrix - column_medians # Отнимаем медиану по столбцам из каждого элемента столбца
print(result_matrix)
```

12) Создать матрицу случайных чисел 5x5. Добавить к ней столбецы максимумов и минимумов по строкам. Получится матрица 5x7.  

```python
import numpy as np
# Создание матрицы 5x5 случайных чисел
matrix = np.random.randint(0, 10, size=(5, 5))
# Добавление столбца максимумов по строкам
max_column = np.max(matrix, axis=1).reshape(-1, 1)
matrix_with_max_column = np.hstack((matrix, max_column))
# Добавление столбца минимумов по строкам
min_column = np.min(matrix, axis=1).reshape(-1, 1)
matrix_with_min_column = np.hstack((matrix_with_max_column, min_column))
print(matrix_with_min_column)
```

14) Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и единицами.

```python
matrix = np.zeros((8, 8), dtype=int)
for i in range(8):
    for j in range(8):
        if (i + j) % 2 == 0:
            matrix[i, j] = 1
print(matrix)
```
